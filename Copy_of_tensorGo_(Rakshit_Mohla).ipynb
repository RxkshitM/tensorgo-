```python
import pandas as pd


df = pd.read_csv('/content/Salary_Data.csv')


numeric_df = df.select_dtypes(include='number')


mean_values = numeric_df.mean()
median_values = numeric_df.median()


mode_df = numeric_df.mode()
if not mode_df.empty:
    mode_values = mode_df.iloc[2]
else:
    mode_values = "No mode available (empty DataFrame)"

std_dev_values = numeric_df.std()
corr_matrix = numeric_df.corr()


print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
print("\nStandard Deviation Values:\n", std_dev_values)
print("\nCorrelation Matrix:\n", corr_matrix)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Salary_Data.csv')


if df.select_dtypes(include='number').empty:
    print("No numerical columns found in the DataFrame. Cannot plot histograms.")
else:

    numeric_df = df.select_dtypes(include='number')


    numeric_df.hist(figsize=(10, 8))
    plt.suptitle('Histograms of Numeric Columns')
    plt.show()


    pd.plotting.scatter_matrix(numeric_df, figsize=(12, 12), alpha=0.8)
    plt.suptitle('Scatter Plot Matrix')
    plt.show()


    numeric_df.plot(kind='line', figsize=(12, 6))
    plt.title('Line Plot of Numeric Columns')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.show()
```
